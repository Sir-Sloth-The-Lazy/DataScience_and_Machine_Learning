### Operators
Operators is python are of following types : 
1. Arithmetic Operators
2. Comparison Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Identity Operators
7. Membership Operators

````python
# 1. Arithmetic Operators
# Variables
a = 15
b = 4

# Addition
print("Addition:", a + b)  

# Subtraction
print("Subtraction:", a - b) 

# Multiplication
print("Multiplication:", a * b)  

# Division
print("Division:", a / b) 

# Floor Division
print("Floor Division:", a // b)  

# Modulus
print("Modulus:", a % b) 

# Exponentiation
print("Exponentiation:", a ** b)
````

````python
# 2. Comparison Operators
a = 13
b = 33

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
````

````python
# 3. Logical Operators
a = True
b = False
print(a and b)
print(a or b)
print(not a)
````

````python
# 4. Bitwise Operators|
a = 10
b = 4

print(a & b) # and operator
print(a | b) # or operator 
print(~a) # not operator
print(a ^ b) # xor operator
print(a >> 2) # right shift operator means divide by 2^x where x is the number of times the bits are shifted
print(a << 2) # left shift operator means multiply by 2^x where x is the number of times the bits are shifted
````

````python
# 5. Assignment Operators
a = 10
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
````

````python
# 6. Identity Operators
# Identity Operators in Python (`is`, `is not`) are used to compare if two variables refer to the *same object* in memory, 
# not just if their values are equal.
#
# Common Misconception:
# Many people confuse `is` with `==`. The `==` operator compares if the values are equal, while `is` checks for object identity 
# (i.e. if both variables point to exactly the same object in memory).
#
# Example:
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True, because their values are equal
print(a is b)  # False, because they are different objects

c = a
print(a is c)  # True, because c references the same object as a

# Lesser-Known Facts:
# - Small integers and some strings are "interned" by Python, which means identical references may point to the same object for efficiency.
x = 256
y = 256
print(x is y)  # True, due to interning

x = 257
y = 257
print(x is y)  # False in many Pythons, because integers > 256 may not be interned

# - For strings, interning sometimes applies:
s1 = "hello"
s2 = "hello"
print(s1 is s2)  # True (usually) due to string interning

s3 = "".join(["he", "llo"])
print(s1 == s3)  # True, same value
print(s1 is s3)  # May be False, not necessarily same object

# - Never use `is` for equality check, only for identity (None check or singleton objects):
foo = None

if foo is None:
    print("foo is None!")  # This is the correct usage

a = 10
b = 20
c = a

print(a is not b)
print(a is c)
````

````python
# 7. Membership Operators
x = 24
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
````

````python
# 8. Ternary Operators
a, b = 10, 20
min = a if a < b else b

print(min)
````

````python
# 9. Operator Precedence
a = 10
b = 20
c = 30    
result = a + b * c
print(result)  # Output: 610
````

````python
# 9. Operator Precedence
a = 10
b = 20
c = 30  
result = (a + b) * c
print(result)  # Output: 900
````

# Type Conversion

There are two types of conversion
- Implicit Type Conversion
- Explicit Type Conversion

## Implicit Type Conversion

Implicit type conversion, also known as coercion, is when Python automatically converts a value from one data type to another without the programmer explicitly requesting it. `Done to protect from loss of data`
For example, during arithmetic operations involving an integer and a float, Python will convert the integer to a float so the operation can continue without error.


````python
num = 1 
num2 = 2.5 
num3 = num + num2 
print(num3)

# In this case, Python automatically converts the integer 1 to a float, allowing the addition to proceed.


````

# Explicit Type Conversion

 Explicit type conversion, also known as type casting, is when the programmer manually changes the data type of a value to another data type. `Might lead to loss of data`
 This is done using built-in functions like int(), float(), str(), etc.
 For example:
 ``` x = "10"
 y = int(x)    # Converts string "10" to integer 10
 print(y + 5)  # Output: 15 
 ```
