Python's `str` class is packed with built-in methods. To make this digestible, Iâ€™ve categorized them by their primary purpose.

Keep in mind that because strings are **immutable**, none of these functions modify the original string; they all return a **new** value.

### 1. Case Transformation Methods

These methods change the visual "casing" of the text.

| Method | Description | Example |
| --- | --- | --- |
| `.upper()` | Converts all characters to uppercase. | `"hi" -> "HI"` |
| `.lower()` | Converts all characters to lowercase. | `"HI" -> "hi"` |
| `.capitalize()` | Capitalizes only the first character. | `"python" -> "Python"` |
| `.title()` | Capitalizes the first character of every word. | `"hello world" -> "Hello World"` |
| `.swapcase()` | Swaps cases (lower to upper and vice-versa). | `"Py" -> "pY"` |
| `.casefold()` | Aggressive lowercase for caseless matching. | `"ÃŸ" -> "ss"` |

---

### 2. Search and Inspection Methods

Use these to find substrings or count occurrences.

| Method | Description | Example |
| --- | --- | --- |
| `.find(sub)` | Returns the lowest index of `sub`. Returns `-1` if not found. | `"abc".find("b") -> 1` |
| `.index(sub)` | Same as `.find()`, but raises `ValueError` if not found. | `"abc".index("z") -> Error` |
| `.count(sub)` | Returns the number of non-overlapping occurrences. | `"banana".count("a") -> 3` |
| `.startswith(prefix)` | Returns `True` if the string starts with the prefix. | `"Hello".startswith("H")` |
| `.endswith(suffix)` | Returns `True` if the string ends with the suffix. | `"file.txt".endswith(".txt")` |

---

### 3. Classification / Boolean Methods

These return `True` or `False` based on the content of the string.

| Method | Returns `True` if... |
| --- | --- |
| `.isalpha()` | All characters are alphabetic (letters). |
| `.isdigit()` | All characters are digits (numbers). |
| `.isalnum()` | All characters are alphanumeric (letters or numbers). |
| `.isspace()` | All characters are whitespace (spaces, tabs, newlines). |
| `.islower()` | All cased characters are lowercase. |
| `.isupper()` | All cased characters are uppercase. |
| `.isprintable()` | The string doesn't contain hidden chars like `\n`. |

---

### 4. Splitting, Joining, and Cleaning

Crucial for data processing and parsing.

| Method | Description | Example |
| --- | --- | --- |
| `.split(sep)` | Splits string into a list based on a separator. | `"a,b,c".split(",") -> ['a', 'b', 'c']` |
| `.join(iterable)` | Joins elements of an iterable into one string. | `"-".join(['a','b']) -> "a-b"` |
| `.strip()` | Removes leading and trailing whitespace. | `"  hi  ".strip() -> "hi"` |
| `.replace(old, new)` | Replaces occurrences of a substring. | `"apple".replace("a", "o") -> "opple"` |
| `.lstrip()` / `.rstrip()` | Removes whitespace from only the Left or Right side. | `"  hi".lstrip() -> "hi"` |

---

### 5. Alignment and Formatting

Used for pretty-printing and UI alignment.

| Method | Description | Example |
| --- | --- | --- |
| `.center(width)` | Centers string within a specific width. | `"hi".center(6) -> "  hi  "` |
| `.ljust(width)` | Left-aligns string. | `"hi".ljust(4) -> "hi  "` |
| `.rjust(width)` | Right-aligns string. | `"hi".rjust(4) -> "  hi"` |
| `.zfill(width)` | Pads the string with zeros on the left. | `"42".zfill(5) -> "00042"` |
| `.format()` | Formats specified values in a string. | `"{0}".format("A")` |
