````python
class LogEntry:
    def __init__(self, raw_log: str):
        self._raw_log = raw_log

        parts = raw_log.split('|')
        self._timestamp = parts[0].strip()

        self._username = None
        self._action = None
        self._status = None
        self._ip_address = None

        for part in parts[1:]:
            key, _, value = part.strip().partition(':')
            key = key.casefold()

            if key == 'user':
                self._username = value
            elif key == 'action':
                self._action = value
            elif key == 'status':
                self._status = value
            elif key == 'ip':
                self._ip_address = value

        # ðŸ”¥ Fail-fast validation
        self._validate()

    def _validate(self):
        # username validation
        if not self._username:
            raise ValueError("Missing username")

        if not self._username.replace('_', '').isalnum():
            raise ValueError("Invalid username format")

        # action validation
        if not self._action:
            raise ValueError("Missing action")

        # status validation
        if not self._status:
            raise ValueError("Missing status")

        if self._status.casefold() not in ("success", "failure"):
            raise ValueError("Invalid status value")

        # IP validation
        if not self._ip_address:
            raise ValueError("Missing IP address")

        if self._ip_address.count('.') != 3:
            raise ValueError("Invalid IP address format")

    # ðŸ”’ Read-only properties

    @property
    def username(self):
        return self._username

    @property
    def action(self):
        return self._action

    @property
    def status(self):
        return self._status

    @property
    def ip_address(self):
        return self._ip_address

    @property
    def timestamp(self):
        return self._timestamp
        
````

LogValidator -- Rule based validation

````python
class LogValidator:
    VALID_ACTIONS = {"login" , "logout" , "purchase"}
    VALID_STATUS = {"success" , "failure"}
    
    @staticmethod
    def validate(entry: LogEntry):
        if entry.action.casefold() not in LogValidator.VALID_ACTIONS:
            raise ValueError("Invalid action")
        
        if entry.status.casefold() not in LogValidator.VALID_STATUS:
            raise ValueError("Invalid status")
        
        if entry.ip_address.count('.') != 3:
            raise ValueError("Invalid IP address")
````

LogAnalyzer â€” aggregation & detection

````python
class LogAnalyzer :
    def __init__(self):
        self._logins_per_user = {}
        self._failure_per_ip = {}

    def process(self, entry: LogEntry):
        if entry.action.casefold() == "login":
            self._logins_per_user[entry.username] = (
                self._logins_per_user.get(entry.username, 0) + 1
        )

        if entry.status.casefold() == "failure":
            self._failure_per_ip[entry.ip_address] = (
                self._failure_per_ip.get(entry.ip_address , 0) + 1
           )
    def suspicious(self, threshold=3):
        return[
            ip for ip, count in self._failure_per_ip.items() if count >= threshold
        ]
 
````

````python
logs = [
    "2026-01-30 10:45:23 | USER:John_Doe | ACTION:LOGIN | STATUS:Success | IP:192.168.1.1",
    "2026-01-30 10:46:10 | USER:alice99 | ACTION:LOGIN | STATUS:FAILURE | IP:192.168.1.1",
    "2026-01-30 10:47:55 | USER:bob | ACTION:LOGOUT | STATUS:Success | IP:10.0.0.5",
    "2026-01-30 10:48:12 | USER:alice99 | ACTION:LOGIN | STATUS:FAILURE | IP:192.168.1.1",
    "2026-01-30 10:49:01 | USER:alice99 | ACTION:LOGIN | STATUS:FAILURE | IP:192.168.1.1",
]
````

````python
analyzer = LogAnalyzer()
for raw_logs in logs:
    try:
        entry = LogEntry(raw_logs)
        LogValidator.validate(entry)
        analyzer.process(entry)
    except ValueError as e:
        print(f"Invalid log entry: {e}")

print("Suspicious IPs:", analyzer.suspicious())
````

## Slicing
slicing syntax is [start:stop:step] where `stop` = 1+ index of the last element we want to include. Just like in the range function : the `end` index is not included



````python
s = "Python"
print(s[0:4])
print(s[:4]) # by default start index is 0
print(s[2:]) # by default end index is the last index
print(s[2:len(s)]) # same as s[2:]
print(s[2::2]) # step size is 2
print(s[::-1]) # reverse the string
print(s[1:-2]) # negative index 

'''
Starting index is included but ending index is not included
'''
````
