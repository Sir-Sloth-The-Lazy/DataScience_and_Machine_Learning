# Slicing 

`string[start : stop : end]`

`start` = starting index of the first character

`end` = 1 + index of the last character. `range` as in range function `end` index is not included

```python
string = "hello"
print(string[1:4])  # 'ell'
print(string[2:len(string)])  # 'llo'
```

```python
var1 = "Monty_Python"
print(var1[-12:-7])  # 'Monty'
```

```python
print(var1[-7:-1])  # 'Pytho'
```

```python
print(var1[-7:-12:-1])  # 'Pytno'
```

```python
letters = "abcde"
print(letters[::2])  # 'ace'
print(letters[1:-2])
```

# Default Values

`string[a : b : c]`
- The default values of `start` and `end` depend upon `step`

- `start` = 0  and `end` = len(string) if `step` > 0

- `start` = len(string) - 1  and `end` = 0 if `step` < 0

```python
# Important
seq = "Hello"
new_seq = seq[-1:0:-1]  # since 0 is the end index it is not included
print(new_seq)  # 'olle'

new_seq = seq[-1:-6:-1] # H -> -1 , e -> -2 , l -> -3 , l -> -4 , o -> -5 hence the last index did not come but
# the string ran out of characters hence all are included in reverse order
print(new_seq)  # 'olleH'
```

```python
# Short hands for common slicing patterns
data = "abcdefghij"
print(data[:5]) # shorthand for data[0:5] which gives string slice from index 0 to 4
print(data[5:]) # shorthand for data[5:len(data)] which gives string slice from index 5 to end
print(data[:])  # shorthand for data[0:len(data)] which gives the whole string
print(data[::]) # shorthand for data[0:len(data):1] which gives the whole string
print(data[:len(data) - 1]) # shorthand for substring from start to second last character of the string
print(data[::3]) # shorthand for every third character from the string starting from index 0
```

```python
# reverse a string
print( "Reversing a string using shorthand : ", data[::-1])  # shorthand for reversing the string
```
